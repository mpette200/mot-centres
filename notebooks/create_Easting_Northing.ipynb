```python
import os
import io

PATH = '/e/DOCS/20_non/05-map-data/data/codepo_aug2020/Data/CSV/'
all_files = os.listdir(PATH)

def parse_line(lines, buffer):
    '''line consists of the following fields:
    postcode, positional_quality_indicator, easting, northing,
     + other non-relevant info.
    
    quality indicator of 90 means it is location info
    so these are skipped
    
    example input:
    "BA1 0AQ",10,375350,164482, ....... other fields
    
    writes output to buffer as space separated list:
    XXXX XXX    Easting    Northing   \n
    0           10         20         29
    '''
    
    for line in lines:
        fields = line.split(',')
        postcode = fields[0][1:-1]
        quality = fields[1]
        easting = fields[2]
        northing = fields[3]

        if quality == '90':
            # skip because location missing
            continue
        
        if ' ' not in postcode:
            # some postcodes are missing spaces
            postcode = postcode[0:4] + ' ' + postcode[4:]
        
        i_begin = buffer.tell()
        buffer.write(' ' * 29)

        buffer.seek(i_begin)
        buffer.write(postcode)

        buffer.seek(i_begin + 10)
        buffer.write(easting)

        buffer.seek(i_begin + 20)
        buffer.write(northing)

        buffer.seek(i_begin + 29)
        buffer.write('\n')
    
    return None

fname = all_files[3]

with open(PATH + fname, newline='') as lines:
    with io.StringIO(newline='\n') as buffer:
        parse_line(lines, buffer)
        data = buffer.getvalue()

output_name = fname.replace('.csv', '.dat')
with open(PATH + output_name, newline='', mode='w') as out:
    out.write(data)
```
